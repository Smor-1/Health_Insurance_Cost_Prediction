```python
import pandas as pd
```

```python
# Load the data from the CSV file
df = pd.read_csv('insurance_01.csv')
df
```

```python
# Create Unique ID number for each row
df['ID'] = range(1, len(df) + 1)
df
```

```python
# Split dataset with ID number and insurance charges into seperate df
info = df.drop(columns=['charges']).iloc[:len(df)//2].reset_index(drop=True)
insurance = df[['ID', 'charges']].iloc[len(df)//2:].reset_index(drop=True)
```

```python
info
```

```python
# Reorder columns to have ID first
info = df[['ID', 'age', 'sex', 'bmi', 'children', 'smoker', 'region']].iloc[:len(df)//2].reset_index(drop=True)
info
```

```python
# export CSVs
info.to_csv('info.csv', index=False)
insurance.to_csv('insurance.csv', index=False)
```
