```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
coverage_df = pd.read_csv("health-protection-coverage.csv")
coverage_df.head()
```

```python
infant_mortality_df = pd.read_csv("child-mortality-vs-health-expenditure.csv", 
                                  usecols = ['Entity',
                                             'Code', 
                                             'Year',
                                            'Mortality rate, under-5 (per 1,000 live births)',
                                           'Current health expenditure per capita, PPP (current international $)'])
infant_mortality_df.head()
```

```python
infant_mortality_df.info()
```

```python
infant_mortality_df = infant_mortality_df.dropna()
```

```python
infant_mortality_df.head()
```

```python
infant_mortality_df.loc[(infant_mortality_df["Year"] > 1995) & infant_mortality_df["Year"] <2020]
```

```python
afghanistan_df = infant_mortality_df.loc[infant_mortality_df["Code"] == "AFG"]
```

```python

x_axis = np.arange(len(afghanistan_df))
tick_locations = [s for s in x_axis]

plt.bar(x_axis, afghanistan_df["Mortality rate, under-5 (per 1,000 live births)"])
plt.xticks(tick_locations, afghanistan_df["Year"], rotation = 'vertical')
plt.title("Infant Mortality Rate in Afghanistan (2002-2019)")
plt.xlabel("Year")
plt.ylabel("Infant Mortality (under 5 per 1000 live births)")
plt.show()
```
