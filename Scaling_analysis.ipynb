```python
# Reused code from Rosas notebook, ensuring data was scaled and if it had an effect on the MSE
# Initial imports
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import hvplot.pandas
from pathlib import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
file_path=("full_data.csv")
insurance_df=pd.read_csv(file_path)
insurance_df.head()
```

```python
# Features with Label encoder
import seaborn as sns
# Transfer data to numeric data
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

label.fit(insurance_df.sex.drop_duplicates())
insurance_df.sex = label.transform(insurance_df.sex)

label.fit(insurance_df.smoker.drop_duplicates())
insurance_df.smoker = label.transform(insurance_df.smoker)

label.fit(insurance_df.region.drop_duplicates())
insurance_df.region = label.transform(insurance_df.region)

insurance_df.head()
```

```python
# Create our features
featured_cols = ['smoker', 'age', 'bmi']  # EDIT HERE select the columns to use as features
X = insurance_df[featured_cols]
```

```python
# Standardize the data with StandardScaler().
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Create our target
y = insurance_df[['charges']]
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test data
y_pred = model.predict(X_test)
```

```python
y_test
```

```python
# Calculate the Mean Squared Error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error: ", mse)
```
