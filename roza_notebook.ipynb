```python
file_path=("financing_healthcare/how-much-we-think-we-spend-on-healthcare-vs-how-much-we-actually-do.csv")
how_much=pd.read_csv(file_path)
how_much=how_much.drop(["Continent"], axis=1)
how_much=how_much.dropna() # there is lots of Nan in this data
how_much.head()
```

```python

how_much2=how_much.drop(columns=["Entity","Year","Population (historical estimates)"])
# Create bar chart
# Set index to cuntry code column
how_much2.set_index('Code', inplace=True)
ax = how_much2.plot(kind='bar')

# Add labels and title
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Double Bar Chart')

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example data
data = {'Category': ['A', 'B', 'C'],
        'Value1': [10, 20, 30],
        'Value2': [15, 25, 35]}
df = pd.DataFrame(data)

# Set index to category column
df.set_index('Category', inplace=True)

# Create bar chart
ax = df.plot(kind='bar')

# Add labels and title
ax.set_xlabel('Category')
ax.set_ylabel('Value')
ax.set_title('Example Double Bar Chart')

# Show plot
plt.show()
```
